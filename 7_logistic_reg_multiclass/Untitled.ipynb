```python
%matplotlib inline 
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits  #it contains 1797 images of size 8*8
```

```python
digits = load_digits()
```

```python
dir(digits) # returns all the properties and methods
```

```python
digits.data[0] # 8*8 array i.e 1st image 
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train,y_test = train_test_split(digits.data ,digits.target ,test_size=0.1)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test , y_test)
```

```python
plt.matshow(digits.images[55])
```

```python
digits.target[55]
```

```python
model.predict([digits.data[55]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test , y_predicted)
cm
```

```python
# confusion matrix visualization
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm , annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
